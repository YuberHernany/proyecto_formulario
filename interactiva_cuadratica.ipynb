# $ax ^{ 2} + bx + c = 0$ 

```python
import numpy as np
import matplotlib.pyplot as plt
from fondo import Cuadratica
```

```python
a = 3 #@param {type:"number"}
b = 3 #@param {type:"number"}
c = 3 #@param {type:"number"}

cuad = Cuadratica(a, b, c)
discriminante = cuad.discriminat()

fig, axes = plt.subplots(1,2)
plt.subplot(121)
ax = plt.gca()
cuad.draw()
dom = cuad.domainNearExtreme()
ax.plot([dom.min(), dom.max()], [0,0])
plt.subplot(122)
ax = plt.gca()
ax.plot([0, 0], [0, discriminante])
ax.set(ylim=(-5,5))
plt.show()

```
